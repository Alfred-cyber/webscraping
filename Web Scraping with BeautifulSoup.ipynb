```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://simple.wikipedia.org/wiki/IPhone'
data = requests.get(url).text
data
```

```python
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
# Creating list with all tables
tables = soup.find_all('table')

#  Looking for the table with the classes 'wikitable' and 'sortable'
table = soup.find('table', class_='wikitable')
table
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['Model', 'Image', 'Storage space', 'Changes', 'Connection', 'Original release date', 'Original Pricing (USD)', 'Minimum OS to sync'])

# Collecting Ddata
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    
    if(columns != []):
        model = columns[0].text.strip()
        image = columns[1].text.strip()
        storage_space = columns[2].text.strip()
        changes = columns[3].text.strip()
        connection = columns[4].text.strip()
        original_release_date = columns[5].text.strip()
        original_pricing = columns[6].text.strip()
        #minimum_os_to_sync = columns[7].text.strip()

        df = df.append({'Model': model,  'Image': image, 'Storage space': storage_space, 'Changes': changes, 'Connection': connection, 'Original release date': original_release_date, 'Original Pricing (USD)' : original_pricing }, ignore_index=True)
        print(df)
    
```

```python
df
```

```python
df[["priceA", "priceB", "priceC"]] = df["Original release date"].str.split(",", expand=True)
df.head()
```

```python
df['priceA'] = df['priceA'].str.replace(r'\D', '')
df['priceB'] = df['priceB'].str.replace(r'\D', '')
df['priceC'] = df['priceC'].str.replace(r'\D', '')
df
```
